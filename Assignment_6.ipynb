```python
# 1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

def fibonacci(n):
    """Displays n number of fibonacci terms"""
    if n <= 1:  
        return 1
    else:  
        return (fibonacci(n-1) + fibonacci(n-2))
    
try:
    terms = int(input("Enter number of terms: "))
    if terms <= 0:  
        raise Exception   
    for i in range(terms):  
        print(fibonacci(i))
except:
    print("Please enter a positive integer")
```

```python
# 2. Write a Python Program to Find Factorial of Number Using Recursion?

def factorial(n):
    """Calculates the factorial of given number"""
    if n <= 1:  
        return 1  
    else:  
        return n * factorial(n-1)  

try:
    num = int(input("Enter a number: "))   
    if num < 0:  
        raise Exception   
    print("The factorial of {0} is {1}".format(num, factorial(num)))
except:
    print("Please enter a positive integer")
```

```python
# 3. Write a Python Program to calculate your Body Mass Index?

def bmi(weight, height):
    """Calculates the BMI for given height(m) and weight(kg)"""
    return weight / (height ** 2)

try:
    wt = float(input("Enter your weight (in kg): "))
    if wt <= 0:  
        raise Exception  
    ht = float(input("Enter your height (in m): "))
    if ht <= 0:  
        raise Exception 
    print("BMI = {}".format(bmi(wt, ht)))
except:
    print("Please enter a positive number")
```

```python
# 4. Write a Python Program to calculate the natural logarithm of any number?

import math

def natural_logarithm(num):
    """Calculates the natural logarithm of the given number"""
    return math.log(num)
try:
    number = float(input("Enter a number: "))
    print("The natural logarithm of {0} is {1}".format(number, natural_logarithm(number)))
except:
    print("Enter a positive real number")
```

```python
# 5. Write a Python Program for cube sum of first n natural numbers?

def cube_sum(n):
    """Calculates sum of cubes of first n natural numbers"""
    sum = 0
    for i in range(n, 0, -1):
        sum += i
    return sum ** 2

try:
    num = int(input("Enter a natural number: "))
    if num <=0:
        raise Exception
    print("The sum of cubes of first {0} natural numbers is {1}".format(num, cube_sum(num)))
except:
    print("Enter a natural number")
```
