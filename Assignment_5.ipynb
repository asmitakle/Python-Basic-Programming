```python
# 1. Write a Python Program to Find LCM?

def gcd(min_num, max_num):
    """Finds GCD of two natural numbers"""
    if max_num < min_num:
        max_num, min_num = min_num, max_num      #max_num contains bigger no of the two entered nos
    
    remainder = max_num % min_num
    if remainder == 0:
        gcd = min_num
        return gcd
    while(remainder != 0):
        max_num, min_num = min_num, remainder
        remainder = max_num % min_num
    gcd = min_num
    return gcd

def lcm(num1, num2):
    """Finds LCM of two natural numbers"""
    return (num1 * num2) / gcd(num1, num2)
    
try:
    num1 = int(input("Enter first number: "))
    if num1 <= 0:
        raise exception
    num2 = int(input("Enter second number: "))
    if num2 <= 0:
        raise exception
    print("LCM of {0} and {1} is {2}".format(num1, num2, lcm(num1,num2)))
except:
    print("Enter positive integer")
```

```python
# 2. Write a Python Program to Find HCF?

def hcf(num1, num2):
    """Finds HCF of two natural numbers"""
    min_num = min(num1, num2)
    max_num = max(num1, num2)
    remainder = max_num % min_num
    if remainder == 0:
        hcf = min_num
        return hcf
    while(remainder != 0):
        min_num, max_num = remainder, min_num
        remainder = max_num % min_num
    hcf = min_num
    return hcf

try:
    num1 = int(input("Enter first number: "))
    if num1 <= 0:
        raise exception
    num2 = int(input("Enter second number: "))
    if num2 <= 0:
        raise exception
    print("HCF of {0} and {1} is {2}".format(num1, num2, hcf(num1,num2)))
except:
    print("Enter positive integer")
```

```python
# 3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

def divider(quo, base):
    """Returns quotient and remainder for the given numbers"""
    return (quo // base, quo % base)

def hexa_digit(num):
    """ Return hexadigit for digits for numbers with base 16""" 
    if num == 10:
        return 'A'
    elif num == 11:
        return 'B'
    elif num == 12:
        return 'C'
    elif num == 13:
        return 'D'
    elif num == 14:
        return 'E'
    elif num == 15:
        return 'F'
    else:
        return str(num)

def convert_rem_list_to_hex_num(lst):
    """Reverses the remainder list to form hexadecimal number"""
    rev_list = lst[::-1]
    converted_hex_num = ''
    
    for digit in rev_list:
        converted_hex_num += digit
        
    if converted_hex_num == "":
        return 0
    
    return converted_hex_num

def base_converter(dec_num, base):
    """Converts a decimal number into a number of the specified base"""

    quo = dec_num
    rem_list = []
    while(quo != 0):
        quo, rem = divider(quo, base)
        
        rem = hexa_digit(rem)
        rem_list.append(rem)
        if quo == dec_num:
            quo = 0

    return convert_rem_list_to_hex_num(rem_list)

try:
    num = int(input("Enter a number: "))
    if num < 0:
        raise Exception
    base = int(input("Enter a base (any natural number >1): "))
    if base <= 1:
        raise Exception
    print(base_converter(num, base))
except:
    print("Enter a valid natural number")
```

```python
# 4. Write a Python Program To Find ASCII value of a character?

def ascii_convert(char):
    return ord(char)

try:
    character = input("Enter a character: ")
    if len(character) != 1:
        raise Exception
    print(ascii_convert(character))
except:
    print("Enter a single character")
```

```python
# 5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

def add(num1, num2):
    return num1 + num2

def sub(num1, num2):
    return num1 - num2

def mul(num1, num2):
    return num1 * num2

def div(num1, num2):
    return num1 / num2

try:
    number1 = float(input("Enter a number: "))
    number2 = float(input("Enter another number: "))
    operator = input("Enter operation to be performed (+,-,*,/): ")
    if operator == '+':
        result = add(number1, number2)
    elif operator == '-':
        result = sub(number1, number2)
    elif operator == '*':
        result = mul(number1, number2)
    elif operator == '/':
        result = div(number1, number2)
    else:
        raise Exception
        
    print("{0} {1} {2} = {3}".format(number1, operator, number2, result))
except ValueError:
    print("Enter a numeric value")
except Exception:
    print("Invalid operator entered")
```
