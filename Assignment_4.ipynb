```python
# 1. Write a Python Program to Find the Factorial of a Number?

def factorial(num):
    """Finds factorial of a given natural number"""
    if num < 0:
        raise Exception("Factorial of negative number does not exist")
    result = 1
    for i in range(num, 1, -1):
        result *= i
    return result

try:
    number = int(input("Enter a positive integer: "))
    print("Factorial is: ", factorial(number))
except Exception:
    print("Please enter a positive integer")
```

```python
# 2. Write a Python Program to Display the multiplication Table?

def mult_table(num):
    """Gives the multiplication table upto 10 for a natural number"""
    if num <= 0:
        raise Exception
    for i in range(num, (num*10) + 1, num):
        yield i
        
try:
    number = int(input("Enter a number: "))
    for x in mult_table(number):
        print(x)
    
except:
    print("Please enter a positive integer")
```

```python
# 3. Write a Python Program to Print the Fibonacci sequence?

def fib(num):
    """Generates fibonacci numbers upto specified number of terms"""
    a = 1
    b = 1
    for i in range(num):
        yield a
        a, b = b, a + b
try:
    terms = int(input("Enter number of terms: "))
    if terms < 0:
        raise Exception
    for x in fib(terms):
        print(x)
except:
    print("Please enter a positive integer")
```

```python
# 4. Write a Python Program to Check Armstrong Number?
#Sum of the individual digits raised to the power of the total number of the digits is equal to the number itself.

def is_armstrong(num):
    """Checks if a given number is an Armstrong number.
    
    Armstrong number:
    Sum of the individual digits raised to the power of the total number of the digits is equal to the number itself.
    """
    if num < 0:
        return False
    num_str = str(num)
    total_digits = len(num_str)
    total = 0
    for i in range(total_digits):
        total += int(num_str[i])**total_digits
    if total == num:
        return True
    else:
        return False

try:
    number = int(input("Enter a number: "))
    if number < 0:
        raise Exception("Armstrong numbers are not negative")
    if (is_armstrong(number)):
        print("{0} is an Armstrong number".format(number))
    else:
        print("{0} is not an Armstrong number".format(number))
    
except:
    print("Please enter a positive integer")
```

```python
# 5. Write a Python Program to Find Armstrong Number in an Interval?

def is_armstrong(num):
    """Checks if a given number is an Armstrong number.
    
    Armstrong number:
    Sum of the individual digits raised to the power of the total number of the digits is equal to the number itself.
    """
    if num < 0:
        return False
    num_str = str(num)
    total_digits = len(num_str)
    total = 0
    for i in range(total_digits):
        total += int(num_str[i])**total_digits
    if total == num:
        return True
    else:
        return False

try:
    start = int(input("Enter start of interval: "))
    if start < 0:
        raise Exception
    end = int(input("Enter end of interval: "))
    if end < 0:
        raise Exception
    ctr = 0
    for number in range(start, end + 1):
        if is_armstrong(number):
            ctr += 1
            print(number)
    print("{0} Armstrong numbers in range {1}-{2}".format(ctr, start, end))
except:
    print("Enter positive integer")
```

```python
# 6. Write a Python Program to Find the Sum of Natural Numbers?

def sum_natural_num(start, end):
    """Finds sum of natural numbers in a given range"""
    return ((start + end) * (end - start + 1)) / 2

try:
    first_num = int(input("Enter positive start number: "))
    if first_num <= 0:
        raise Exception
    last_num = int(input("Enter positive end number: "))
    if last_num <= 0:
        raise Exception
    print("Sum of numbers in given range is {0}".format(sum_natural_num(first_num, last_num)))
except:
    print("Enter positive integer")
```
