```python
def populate_list():
    ls = []
    num  = 0
    try:
        while True:
            num = int(input("Enter any number to continue : "))
            ls.append(num)
            continue
    except:
        print("Entered elements:", ls)
        
    return ls
```

```python
# 1. Write a Python Program to find sum of array?

"""from functools import reduce

def add(a, b):
    return a+b

def array_sum(array):
    return reduce(add, array)"""

def array_sum(array):
    total = 0
    for num in array:
        total += num
    return total

my_list = populate_list()
    
print("Sum of numbers is", array_sum(my_list))
```

```python
# 2. Write a Python Program to find largest element in an array?

def array_max(array):
    max_num = array[0]
    for num in array:
        if num > max_num:
            max_num = num
    return max_num

my_list = populate_list()

print(array_max(my_list))
```

```python
# 3. Write a Python Program for array rotation?

def array_rotate(array, rotn):
    rotd_list = []
#     for i in range(rotn):
#         rotd_list.append(array.pop(0))
    rotd_list = [array.pop(0) for i in range(rotn)]
    array.extend(rotd_list)
    return array
    
my_list = populate_list()
rotn = int(input("Enter number of elements to be rotated: "))

print("Rotated array:", array_rotate(my_list, rotn))
```

```python
# 3. Using Juggling algorithm ref: geeksforgeeks

def printArray(arr, size):
    for i in range(size):
        print ("% d" % arr[i], end =" ")

def gcd(a, b):
    if b == 0:
        return a;
    else:
        return gcd(b, a % b)

def leftRotate(arr, d, n):
    d = d % n
    g_c_d = gcd(d, n)
    for i in range(g_c_d):
         
        # move i-th values of blocks
        temp = arr[i]
        j = i
        while 1:
            k = j + d
            if k >= n:
                k = k - n
            if k == i:
                break
            arr[j] = arr[k]
            j = k
        arr[j] = temp
        
arr = populate_list()
n = len(arr)
#d = 2
d = int(input("Enter number of elements to be rotated: "))

leftRotate(arr, d, n)
printArray(arr, n)
```

```python
# 4. Write a Python Program to Split the array and add the first part to the end?

def array_split(array, pos):
    split_list = []
#     for i in range(pos):
#         split_list.append(array.pop(0))
#     array.extend(split_list)
    split_list = array[pos:]+array[0:pos]
    return split_list
    
my_list = populate_list()
pos = int(input("Enter where array should be split: "))

print("New array:", array_split(my_list, pos))
```

```python
# 5. Write a Python Program to check if given array is Monotonic?

def chk_sign(elem):
    if elem > 0:
        return lambda x : x > 0
    elif elem < 0:
        return lambda x : x < 0

def isMonotonic(array):
    if len(array) == 1:
        return True
    diff_list = [array[i] - array[i-1] for i in range(1, len(array))]

    filtered = list(filter(chk_sign(diff_list[0]), diff_list))
    
    if len(filtered) == len(diff_list):
        return True
    else:
        return False
    
    #     if diff_list[0] > 0:
    #         for num in diff_list:
    #             if num < 0 or num == 0:
    #                 return False
    #         else:
    #             return True
    #     elif diff_list[0] < 0:
    #         for num in diff_list:
    #             if num > 0 or num == 0:
    #                 return False
    #         else:
    #             return True
    
my_list = populate_list()
print("The list is monotonic:", isMonotonic(my_list))
```

```python
# 5. Alternate method

def isMonotonic2(array):
    if len(set(array)) == len(array):
        if sorted(array) == array or sorted(array, reverse = True) == array:
            return True
        else:
            return False
    else:
        return False

my_list = populate_list()
print("The list is monotonic:", isMonotonic2(my_list))
```
