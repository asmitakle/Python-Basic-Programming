```python
# 1. Write a Python program to convert kilometers to miles?

def kilometer_to_mile(kilometer):
    return kilometer * 0.6214
try:
    kilometer = float(input("Enter distance in kilometers: "))
    mile = kilometer_to_mile(kilometer)
    print("Miles = ", mile)
except:
    print("Please enter a numeric value")
```

```python
# 2. Write a Python program to convert Celsius to Fahrenheit?

def celsius_to_fahrenheit(celsius):
    return ((celsius * 9) / 5) + 32
try:
    celsius = float(input("Enter temperature in celsius: "))
    fahrenheit = celsius_to_fahrenheit(celsius)
    print("Fahrenheit = ", fahrenheit)
except:
    print("Please enter a numeric value")
```

```python
# 3. Write a Python program to display calendar?

import calendar
try:
    year = int(input("Enter the year: "))
    print(calendar.calendar(year))
except:
    print("Please enter a integer value")
```

```python
# 4. Write a Python program to solve quadratic equation?

def quadratic_roots(a, b, c):
    x1 = (-b + (b**2 - (4*a*c))**(1/2)) / (2*a)
    x2 = (-b - (b**2 - (4*a*c))**(1/2)) / (2*a)
    return (x1,x2)
try:    
    a = int(input("Enter coefficient of x^2: "))
    b = int(input("Enter coefficient of x: "))
    c = int(input("Enter constant term: "))

    (root1, root2) = quadratic_roots(a, b, c)
    print("Roots of the function are: ", root1, " and ", root2)
    
except:
    print("Please enter a integer value")
```

```python
# 5. Write a Python program to swap two variables without temp variable?

def swap(a, b):
    a, b = b, a
    return (a,b)
    
var1 = input("Enter value for variable 1: ")
var2 = input("Enter value for variable 2: ")
print("Values entered are: {0} and {1}".format(var1, var2))
var1, var2 = swap(var1, var2)
print("Swapped values are: {0} and {1}".format(var1, var2))
```
